Q1 (25%) is the first quartile, also known as the lower quartile. It’s the middle number between the smallest number and the median of the dataset.
Q2 (50%) is the second quartile or median. It divides the data into two halves.
Q3 (75%) is the third quartile, also known as the upper quartile. It’s the middle value between the median and the highest value of the dataset.

```python
tips_df.groupby('sex')['total_bill'].mean()
```

```python
sns.barplot(x='sex',y='tip',data=tips_df)
```

```python
sns.stripplot(x='day',y='total_bill',data = tips_df)
```

```python
sns.stripplot(x='day',y='total_bill',hue = 'sex',data=tips_df)
```

```python
sns.stripplot(x='day',y='total_bill',hue = 'sex',dodge=True,data=tips_df)
```

```python
#swarmplot()-doesn't let individual data point to get merge, it represents each data pont separately.
sns.swarmplot(x='day',y='total_bill',hue = 'sex',dodge=True,data=tips_df)
```
