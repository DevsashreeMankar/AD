```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([1,2,3,4,5,6])
y1 = x1**2
```

```python
#labelling the graph
plt.plot(x1,y1,'r')
plt.xlabel("Numbers")
plt.ylabel("Square of Numbers")
plt.title("Numbers vs Square of Numbers")
```

```python
plt.subplot(2,3,1)
plt.plot(x1,y1,'r--')
plt.subplot(2,3,5)
plt.plot(y1,x1,'r*-')
plt.subplot(2,3,3)
plt.plot(x1,y1,'r+')
```

```python
#scatter plot
a = np.array([1,5,8,6])
b = np.array([8,3,5,9])
```

```python
plt.scatter(a,b,marker='s',color='red')
```

```python
plt.plot(a,b)
```

```python
#bar plot
a = [1990,1991,1992,1993,1994,1995,1996]
b = [10,20,30,40,50,60,70]
plt.bar(a,b,color='green',width=0.4)
plt.xlabel("a")
plt.ylabel("b")
plt.title("a vs b")
```

```python
#legend graph
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0,0,1,1])
ax.plot(x1,x1**2,label='Square of Numbers')
ax.plot(x1,x1**3,label='Cube of Numbers')
ax.legend()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
#left,bottom,width,height
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])
axes1.plot(x1,y1,'b')
#setting values manually because the graph is being plotted by using users input
axes1.set_xlabel("Numbers")
axes1.set_ylabel("Square of Numbers")
axes2.plot(y1,x1,'r--')
axes2.set_xlabel("Square of Numbers")
axes2.set_ylabel("Numbers")
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x1,x1**2,label='Square of Numbers')
ax.plot(x1,x1**3,label='Cube of Numbers')
ax.legend(loc=0)
```

loc=0 :by default vaue
(best suitable as per the graph)
loc=1 :right top corner
loc=2 :left top corner
loc=3 :left bottom corner
loc=4 :right bottom corner

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x1,x1+1,'r',lw=1,ls='-',marker='+',markersize=10) #lw:linewidth,ls:linestyle
ax.plot(x1,x1+2,'r',lw=2,ls='--',marker='o',markersize=4)
ax.plot(x1,x1+3,'r',lw=3,ls='-',marker='s',markersize=6)
ax.plot(x1,x1+4,'r',lw=4,ls='--',marker='*',markersize=8)
```
